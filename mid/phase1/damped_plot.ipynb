```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('results.csv')

display(data)

print(data.columns)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(data['t'], data['euler_x'], label='Euler (Drift)', linestyle='--')
plt.plot(data['t'], data['rk4_x'], label = "RK4 (Stable)", linewidth=2)
plt.title("Euler vs RK4 : Spring Mass System")
plt.xlabel("Time")
plt.ylabel("Position (x)")
plt.legend()
plt.grid(True)
plt.show()
```
